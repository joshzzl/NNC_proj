```python
import numpy as np
import numpy.random as random
import itertools
import warnings
from math import exp
from model import Model
warnings.filterwarnings('ignore')
```

```python
# Input Data X
X = [[1,1], [1,-1], [-1,1], [-1,-1]]

# Input target Y
Y = [-1,1,1,-1]

X = np.array(X)
Y = np.array(Y)
```

```python
w01 = np.array([[-0.3378,0.2771,0.2859,-0.3329],
                [0.1970,0.3191,-0.1448,0.3594],
                [0.3099,0.1904,-0.0347,-0.4861]])
w12 = np.array([[-0.1401], [0.4919], [-0.2913], \
                [-0.3979], [0.3581]])
```

```python
print(w01.shape)
print(w12.shape)
```

### Experiment with quadratic cost function

```python
model = Model(N0=2, N1=4, N2=1, x0=1.0, lr=0.2, tol=0.05, \
              zeta=1, epoch_thres=5000, cost='L2', transfer='tansig')
```

```python
# put in the preset weights, train for 1 epoch
model.train(X, Y, epochs=1, input_weights=True, w01=w01, w12=w12)
```

```python
print(model.predict(X))
print(model.stop_epoch)
print(model.w01)
print(model.w12)
```

```python
# put in the preset weights, 
# train until converge or reach the epoch threshold
model.train(X, Y, epochs=-1, input_weights=True, w01=w01, w12=w12)
```

```python
print(model.predict(X))
print(model.stop_epoch)
print(model.w01)
print(model.w12)
```

```python
# randomize the weights with zeta
model.train(X, Y, epochs=-1, input_weights=False)
```

```python
print(model.predict(X))
print(model.stop_epoch)
print(model.w01)
print(model.w12)
```
